### **1. Import Libraries and Dataset**

```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv("heart.csv")
df = df0.copy()
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df["thal"] = df["thal"].replace(0, np.nan)
df["thal"].fillna(2, inplace = True)
df["thal"] = df["thal"].astype("int")
```

### **Logistic Regression**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', LogisticRegression())
])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

log_reg_df = pd.DataFrame({"Model": "Logistic Regression Classifier",
              "Accuracy": accuracy,
              "Precision": precision,
              "Recall": recall,
              "F1-Score": f1},
            index = [0])
log_reg_df
```

### **Support Vector Machines**

```python
from sklearn.svm import SVC
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', SVC())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

svc_df = pd.DataFrame({"Model": "Support Vector Classifier",
              "Accuracy": accuracy,
              "Precision": precision,
              "Recall": recall,
              "F1-Score": f1},
            index = [0])
svc_df
```

### **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', RandomForestClassifier())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

rf_df = pd.DataFrame({"Model": "Random Forest Classifier",
              "Accuracy": accuracy,
              "Precision": precision,
              "Recall": recall,
              "F1-Score": f1},
            index = [0])
rf_df
```
